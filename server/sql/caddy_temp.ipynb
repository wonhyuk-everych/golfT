```python
import csv

def csv_to_sql(csv_text: str) -> str:
    reader = csv.reader(csv_text.strip().splitlines())
    values = []
    def esc(s: str) -> str:
        return s.replace("'", "''").strip()

    for row in reader:
        if not row or len(row) < 4:
            continue
        caddy_num = int(row[0].strip())
        name = esc(row[1])
        nickname = esc(row[2])
        golf_name = esc(row[3])
        values.append(f"({caddy_num}, '{name}', '{nickname}', '{golf_name}')")

    if not values:
        return "-- No valid rows"
    return "INSERT INTO caddy_temp(caddy_num, name, nickname, golf_name) VALUES\n  " + ",\n  ".join(values) + ";"

print(csv_to_sql(text))
```
