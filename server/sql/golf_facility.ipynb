```python
import pandas as pd
import re

df = pd.read_excel('apl_golf_data.xlsx', sheet_name='골프장_입력')

# 제일 끝에 있는 열의 이름을 가져오기
last_column_name = df.columns[-1]

print(last_column_name)

# 열 이름에 해당 되는 값 가져오기
last_column_values = df[last_column_name].str.replace(r"\s*,\s*", "\n", regex=True)
#print(last_column_values)

# df 돌면서 UPDATE golf_course SET facility_info = #여기에 값# WHERE course_idx = #8부터 시작# 만들어줘
for i in range(len(df)):
    sql = f"UPDATE golf_course SET facility_info = '{last_column_values[i]}' WHERE course_idx = {i+8};"
    print(sql)


```

```python


# 모든 값을 텍스트로 모두 가져오기
last_column_values_text = last_column_values.to_list()

print(last_column_values_text)

# \n 문자열 제거, ' 제거
last_column_values_text = [value.replace("\n", "") for value in last_column_values_text]

print(last_column_values_text)

# for 문 돌면서 텍스트를 , 구분한뒤 각 단어들을 리스트에 중복없이 넣어줘
word_list = []
for value in last_column_values_text:
    words = value.split(",")
    for word in words:
        word = word.strip()
        if word not in word_list:
            word_list.append(word)

print(word_list)
```

```python
# 리스트의 각 단어들에서 띄어 쓰기 제거 한뒤 중복 제거
word_list = list(set([word.replace(" ", "") for word in word_list]))
print(len(word_list))

print(word_list)
```
