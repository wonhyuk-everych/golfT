```python
import pandas as pd

df = pd.read_excel('facility.xlsx')

# 첫번째 열 가져오기
first_column = df.iloc[:, 2]
```

```python
# NaN 값 제거
first_column = first_column.dropna()
```

```python

# 쿼리로 만들기 update hotel_info set hotel_facility_info = #값# where hotel_idx = #73 부터 하나씩 증가#
start_idx = 74
for i, val in enumerate(first_column):
    idx = start_idx + i
    # 따옴표 이스케이프
    sval = val.replace("'", "''")
    print(f"UPDATE hotel_info SET hotel_room_facility_info = '{sval}' WHERE hotel_idx = {idx};")
```
